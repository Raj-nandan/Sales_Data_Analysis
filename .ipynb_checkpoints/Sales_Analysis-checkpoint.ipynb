# importing packages

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter
```

# merging all the data of csv file in one

```python
files = [file for file in os.listdir('SalesAnalysis/Sales_Data/')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("SalesAnalysis/Sales_Data/"+ file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index = False)
```

# reading data

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

# cleaning data

```python
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()

all_data = all_data.dropna(how= 'all')
all_data.head()
```

# Fixing error of 'or' data in order date

```python
or_pd = all_data[all_data['Order Date'].str[0:2] == 'Or']
or_pd.head()

all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

# fixing data types of some column

```python
#to int
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
#to float
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

# adding more info to dataset

```python
#months column
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')

#salses Column
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']

#city column
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
    
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

# Time for some analysis over the cleaned data

# Q) What was the month for highest sales and How much was the sales

```python
results = all_data.groupby('Month').sum(numeric_only=True)
```

```python
months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.show()
```

# Q) Which city has the highest sales

```python
results = all_data.groupby('City').sum(numeric_only=True)
results
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation= 'vertical', size = 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City names')
plt.show()
```

# Q) What time should we display advertisement to maximize likelihood of customer's byuing product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
plt.xlabel('Hours')
plt.ylabel('Numbers of orders')

plt.show()
```

# Q) What products are more often sold together

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
#creating new column for grouped product
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID' , 'Grouped']].drop_duplicates()

df.head()
```

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```

# Q) What product sold the most and why do you think it sold the most

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group['Quantity Ordered'].sum()

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation= 'vertical', size = 8)
plt.xlabel('Products')
plt.ylabel('Quantity')
plt.show()
```

```python
prices = all_data.groupby('Product').mean(numeric_only=True)['Price Each']


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered)
ax2.plot(products, prices, 'g-')


ax1.set_xlabel('products')
ax1.set_ylabel('quantity', color='g')
ax2.set_ylabel('prices', color='b')
ax1.set_xticks(range(len(products)))
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)

plt.show()
```
